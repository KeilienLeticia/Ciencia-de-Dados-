```python
import pandas as pd
import pyodbc 

server = 'LETICIA' # Substitua pelo nome do servidor SQL Server
database = 'Python'  # Substitua pelo nome do banco de dados
conexaoDB = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};'
                      f'SERVER={server};'
                      f'DATABASE={database};'
                      'Trusted_Connection=yes;')

cursor = conexaoDB.cursor()   # criando cursor de comando '
```

```python
Produto_Categoria = '''
select 
p.Nome, 
p.Id_Category,
i.order_id,
i.quantity,
i.total_price
from  [Produtos]     as p
left join [items]     as i on p.id = i.id
'''
Resultado = pd.read_sql(Produto_Categoria , conexaoDB)
display(Resultado)
```

```python
Categoria_ordens = '''
select 
o.created_at,
o.customer_id,
o.pagamento
from  [dbo].[Categoria] as c
left join [dbo].[Ordens] as o on c.id=o.id
'''
Cat_ord=pd.read_sql(Categoria_ordens , conexaoDB)
display(Cat_ord)
```

```python
Juncao = '''
select 
p.Nome, 
p.Id_Category,
i.order_id,
i.quantity,
i.total_price,
c.Nome,
cli.created_at, 
cli.last_name,
cli.cell_phone
from  [Produtos]     as p
left join [items]     as i on p.id = i.id
left join [Categoria] as c on p.id = c.id
left join [Clientes]  as cli on c.id = cli.id

'''
Juncao_tabelas = pd.read_sql(Juncao , conexaoDB)
Juncao_tabelas.head(10)
```
