```python

import pandas as pd
import pyodbc 

server = 'LETICIA' # Substitua pelo nome do servidor SQL Server
database = 'AdventureWorksDW2017'  # Substitua pelo nome do banco de dados
conexaoDB = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};'
                      f'SERVER={server};'
                      f'DATABASE={database};'
                      'Trusted_Connection=yes;')

cursor = conexaoDB.cursor()   # criando cursor de comando 
```

```python
Query = """
select * from [dbo].[DimPromotion]
where DiscountPct >0.1

"""
```

```python
Vendas = pd.read_sql(Query,conexaoDB)
Vendas.head(10)
```
